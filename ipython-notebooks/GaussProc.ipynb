```python
# Gaussian processes in action
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib;matplotlib.rcParams['figure.figsize'] = (8,5)
```

```python
import GPy
import numpy as np
```

```python
# chose appropriate Kernel 
kernel = GPy.kern.RBF(input_dim = 1, variance = .75, lengthscale=2.)
_ = kernel.plot(ax=plt.gca())
```

```python
X = np.random.uniform(-3., 3., (20, 1))       # pick some random points
Y = np.sin(X) + np.random.randn(20, 1)*0.05   # apply some function to the points
plot(X, Y, '*')
xs = np.linspace(-3., 3., 200)
plot(xs, np.sin(xs))
```

```python
# get a model
model = GPy.models.GPRegression(X, Y, kernel)
```

```python
model.plot()
plot(X, Y, '*')
xs = np.linspace(-3., 3., 200)
plot(xs, np.sin(xs))
```

```python
model.optimize_restarts(num_restarts=10)
model.plot()
plot(X, Y, '*')
xs = np.linspace(-3., 3., 200)
plot(xs, np.sin(xs))
```
