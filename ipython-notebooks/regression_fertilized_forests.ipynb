```python
# Data
path_to_csv = "Costs50.csv"
data = np.genfromtxt(path_to_csv, dtype=float, delimiter=';')
power50 = data[:,0]
costs50 = data[:,1]
# For training.
X = np.ascontiguousarray(np.atleast_2d(power50).T)
Y = np.ascontiguousarray(np.atleast_2d(costs50).T)
```

Use double as datatype, configure forest for regression.

```python
soil = fertilized.Soil('d', 'd', 'd', fertilized.Result_Types.regression)
forest = soil.StandardRegressionForest(1, # Number of features
                                       5, # Maximum depth
                                       1, # Tested features per node
                                       3, # Tested thresholds
                                       10, # # trees
                                       5, # Min samples per leaf
                                       10  # Min samples per split
                                       )
forest.fit(X, Y)
```

Plot the interval [0, 600] with 600 samples.

```python
def create_plot(forest, title_add='', marked_points=None, plot_grid=(2,1)):
    assert plot_grid[0] * plot_grid[1] >= 2
    # Define the x values for prediction input.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    # Predict the y values based on X_test
    prediction = forest.predict(X_test)
    y_upper_conf = []
    y_lower_conf = []
    for x_ind, x in enumerate(X_test):
        interval = scipy.stats.norm.interval(0.95, loc=prediction[x_ind, 0], scale=np.sqrt(prediction[x_ind, 1]))
        y_lower_conf.append(interval[0])
        y_upper_conf.append(interval[1])
    # Create the plot.
    # Data, prediction and mean confidence intervals
    plt.subplot(plot_grid[0], plot_grid[1], 1)
    plt.title('Data, prediction and 95% mean confidence interval' + title_add)
    plt.plot(X, Y, label='Observations', linestyle='None', **markerparams)
    plt.plot(X_test, prediction[:, 0], label='Prediction')
    plt.plot(X_test, y_upper_conf, label='95% upper confidence border')
    plt.plot(X_test, y_lower_conf, label='95% lower confidence border')
    if not marked_points is None:
        plt.plot(X[marked_points], Y[marked_points], label='Used points',
                 marker='x', markeredgecolor='red', markeredgewidth=1,
                 linestyle='None')
    plt.xlabel('$P$')
    plt.ylabel('Cost')
    plt.legend(loc='lower right')
    # Data, prediction and sigma.
    plt.subplot(plot_grid[0], plot_grid[1], 2)
    plt.title(r'Data, prediction and $\sigma$' + title_add)
    plt.plot(X, Y, label='Observations', **markerparams)
    plt.plot(X_test, prediction[:, 0], label='Prediction')
    if not marked_points is None:
        plt.plot(X[marked_points], Y[marked_points], label='Used points',
                 marker='x', markeredgecolor='red', markeredgewidth=1,
                 linestyle='None')
    plt.plot(X_test, 10.*np.sqrt(prediction[:, 1]), label=r'$10\cdot\sigma$')
    conv_var = np.convolve(prediction[:, 1], signal.gaussian(50, 10.), mode='same')
    plt.plot(X_test, np.sqrt(conv_var), label=r'$\sigma$ gauss filtered')
    plt.xlabel('$P$')
    plt.ylabel('Cost')
    plt.legend(loc='upper left')
plt.figure(figsize=(15, 10))
create_plot(forest)
```

## Active learning

```python
from IPython import display
import time
from sklearn.metrics import mean_squared_error
def showAndRedraw(delay):
    r"""Zeigt den momentanen Plot and und wartet `delay` Sekunden
    bis zum weiterrechnen. Der Plot wird beim nächsten Aufruf
    ersetzt."""
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(delay)
    plt.gcf().clear()
```

```python
n_initial_points = 5
allow_readdition = False
use_gaussian_convolved_for_resampling=False
plt.figure(figsize=(15, 20))
sample_points = np.round(np.linspace(0, power50.shape[0]-1, num=n_initial_points)).astype('int')
errors = []
errors_relative = []
errors_sample_counts = []
for add_point_idx in xrange(45):
    forest = soil.StandardRegressionForest(1, # Number of features
                                           5, # Maximum depth
                                           1, # Tested features per node
                                           10, # Tested thresholds
                                           200, # # trees
                                           2, # Min samples per leaf
                                           4  # Min samples per split
                                           )
    X_sampled = X[sample_points, :]
    Y_sampled = Y[sample_points, :]
    forest.fit(X_sampled, Y_sampled)
    create_plot(forest,
                ' using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                  sample_points.shape[0]),
                sample_points,
                (3, 1))
    # Measure the error.
    error = mean_squared_error(Y, forest.predict(X)[:, 0])
    errors_sample_counts.append(np.unique(sample_points).shape[0])
    errors.append(error)
    errors_relative.append(float(errors_sample_counts[-1])/ error)
    plt.subplot(3,1,3)
    plt.title('Error using %d samples (%d total)' % (np.unique(sample_points).shape[0],
                                                     sample_points.shape[0]))
    plt.xlabel('Number of sample points')
    plt.ylabel('MSE')
    plt.plot(errors_sample_counts, np.array(errors), label='MSE')
    #plt.plot(errors_sample_counts, errors_relative, label='MSE/npoints')
    plt.legend(loc='upper left')
    showAndRedraw(0.2)
    # Find the most promising point and resample.
    X_test = np.linspace(0, np.max(X), num=600)
    X_test = np.atleast_2d(X_test).T
    prediction = forest.predict(X_test)
    suggested_x = None
    conv_var = np.convolve(prediction[:, 1], signal.gaussian(50, 10.), mode='same')
    candidates = range(prediction.shape[0])
    while suggested_x is None:
        if use_gaussian_convolved_for_resampling:
            suggested_x_idx = np.argmax(conv_var[candidates])
        else:
            suggested_x_idx = np.argmax(prediction[candidates, 1])
        suggested_x_idx = candidates[suggested_x_idx]
        suggested_x = X_test[suggested_x_idx]
        # Get closest known powervalue.
        closest_x_idx = np.argmin((power50-suggested_x)**2)
        if closest_x_idx in sample_points and not allow_readdition:
            candidates = [candidate for candidate in candidates if candidate != suggested_x_idx]
            suggested_x = None
    sample_points = np.hstack((sample_points, closest_x_idx))
```

### Evaluation measure

The evaluation measure should capture, how good the learned piecewise linear function approximates the original data. The classical MSE could be one measure to use here for a single error approximation.

I suggest to use a weighted error measure for the entire system: for $n$ used sample points, define the error as
$$ \frac{1}{\sum \frac{1}{n}}\cdot \sum_n \frac{1}{n}\cdot \mbox{MSE}_n.$$

This is a weighted mean over the MSE errors, with the weights corresponding to the reciprocal number of sample points used.


```python
def WMSE(errors, errors_sample_counts):
    return 1./np.sum(1./np.array(errors_sample_counts))*np.sum((1./np.array(errors_sample_counts))*np.array(errors))
WMSE(errors, errors_sample_counts)
```

## How good can it get?

```python
def train_evaluate(maxdepth, threshold, trees, leaf_samples, gain_threshold, use_gaussian):
    n_initial_points = 5
    allow_readdition = False
    use_gaussian_convolved_for_resampling=use_gaussian
    plt.figure(figsize=(15, 20))
    sample_points = np.round(np.linspace(0, power50.shape[0]-1, num=n_initial_points)).astype('int')
    errors = []
    errors_relative = []
    errors_sample_counts = []
    for add_point_idx in xrange(51):
        forest = soil.StandardRegressionForest(1, # Number of features
                                               maxdepth, # Maximum depth
                                               1, # Tested features per node
                                               threshold, # Tested thresholds
                                               trees, # # trees
                                               leaf_samples, # Min samples per leaf
                                               2*leaf_samples,  # Min samples per split
                                               min_gain_threshold=gain_threshold
                                               )
        X_sampled = X[sample_points, :]
        Y_sampled = Y[sample_points, :]
        forest.fit(X_sampled, Y_sampled)
        # Measure the error.
        error = mean_squared_error(Y, forest.predict(X)[:, 0])
        errors_sample_counts.append(np.unique(sample_points).shape[0])
        errors.append(error)
        if n_initial_points + add_point_idx < X.shape[0]:
            # Find the most promising point and resample.
            X_test = np.linspace(0, np.max(X), num=600)
            X_test = np.atleast_2d(X_test).T
            prediction = forest.predict(X_test)
            suggested_x = None
            conv_var = np.convolve(prediction[:, 1], signal.gaussian(50, 10.), mode='same')
            candidates = range(prediction.shape[0])
            while suggested_x is None:
                if use_gaussian_convolved_for_resampling:
                    suggested_x_idx = np.argmax(conv_var[candidates])
                else:
                    suggested_x_idx = np.argmax(prediction[candidates, 1])
                suggested_x_idx = candidates[suggested_x_idx]
                suggested_x = X_test[suggested_x_idx]
                # Get closest known powervalue.
                closest_x_idx = np.argmin((power50-suggested_x)**2)
                if closest_x_idx in sample_points and not allow_readdition:
                    candidates = [candidate for candidate in candidates if candidate != suggested_x_idx]
                    suggested_x = None
            sample_points = np.hstack((sample_points, closest_x_idx))
    # Get the total error value.
    total_error = WMSE(errors, errors_sample_counts)
    return total_error
```

```python
best_params = {}
best_value = np.infty # Minimize
maxdepth_values = (2,3,4,5,7,9)
threshold_values = (1,3,5,7,9)
trees_values = (2,8,16,32,64,128,256,512)
leaf_samples_values = (2,3,4,5)
gain_threshold_values = (0., 1E-7)
use_gaussian_values = (True, False)
for maxdepth in maxdepth_values:
    for threshold in threshold_values:
        for trees in trees_values:
            for leaf_samples in leaf_samples_values:
                for gain_threshold in gain_threshold_values:
                    for use_gaussian in use_gaussian_values:
                        error = train_evaluate(maxdepth, threshold, trees, leaf_samples, gain_threshold, use_gaussian)
                        if error < best_value:
                            best_value = error
                            best_params = {'maxdepth': maxdepth,
                                           'threshold': threshold,
                                           'trees': trees,
                                           'leaf_samples': leaf_samples,
                                           'gain_threshold': gain_threshold,
                                           'use_gaussian': use_gaussian}
                            print best_params, best_value
```
