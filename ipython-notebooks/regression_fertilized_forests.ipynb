# Regression with the fertilized forest library

```python
# Imports
import os
import sys
sys.path.insert(0, r'C:\thesis\pythonWorkspace\VisualStudio\fertilized-forests\pyfertilized')
import numpy as np
np.random.seed(0)
import matplotlib.pylab as plt
import scipy.stats
import fertilized
```

```python
# Data
path_to_csv = "Costs50.csv"
data = np.genfromtxt(path_to_csv, dtype=float, delimiter=';')
power50 = data[:,0]
costs50 = data[:,1]
# For training.
X = np.ascontiguousarray(np.atleast_2d(power50).T)
Y = np.ascontiguousarray(np.atleast_2d(costs50).T)
```

```python
markerparams = {'marker':'o',
                'markerfacecolor':'None',
                'markeredgecolor':'blue',
                'markersize':6,
                'markeredgewidth':1}
```

```python
plt.plot(power50, costs50, **markerparams)
plt.xlabel('$P$')
plt.ylabel('Cost')
```

Use double as datatype, configure forest for regression.

```python
soil = fertilized.Soil('d', 'd', 'd', fertilized.Result_Types.regression)
forest = soil.StandardRegressionForest(1, # Number of features
                                       5, # Maximum depth
                                       1, # Tested features per node
                                       3, # Tested thresholds
                                       10, # # trees
                                       5, # Min samples per leaf
                                       10  # Min samples per split
                                       )
forest.fit(X, Y)
```

Plot the interval [0, 600] with 600 samples.

```python
# Define the x values for prediction input.
X_test = np.linspace(0, 600, num=600)
X_test = np.atleast_2d(X_test).T
# Predict the y values based on X_test
prediction = forest.predict(X_test)
```

```python
y_upper_conf = []
y_lower_conf = []
for x_ind, x in enumerate(X_test):
    interval = scipy.stats.norm.interval(0.95, loc=prediction[x_ind, 0], scale=np.sqrt(prediction[x_ind, 1]))
    y_lower_conf.append(interval[0])
    y_upper_conf.append(interval[1])
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
plt.title('Data, prediction and scaled 95% confidence interval')
plt.plot(X, Y, label='Observations', linestyle='None', **markerparams)
plt.plot(X_test, prediction[:, 0], label='Predictions')
plt.plot(X_test, y_upper_conf, label='95% upper confidence border')
plt.plot(X_test, y_lower_conf, label='95% lower confidence border')
plt.xlabel('$P$')
plt.ylabel('Cost')
plt.legend(loc='upper left')

plt.subplot(2, 1, 2)
plt.title(r'Data, prediction and $\sigma$')
plt.plot(X, Y, label='Observations', **markerparams)
plt.plot(X_test, 10.*np.sqrt(prediction[:, 1]), label=r'$10\cdot\sigma$')
plt.xlabel('$P$')
plt.ylabel('Cost')
plt.legend(loc='upper left')
```
